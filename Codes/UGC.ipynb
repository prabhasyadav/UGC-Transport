```python
import numpy as np
import pylab as pl 
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
adv1 = np.genfromtxt('fig21.csv', delimiter=',')  
diff1 =np.genfromtxt("fig22.csv", delimiter=",") 
diff2 =np.genfromtxt("fig23.csv", delimiter=",") 
diff3 =np.genfromtxt("fig24.csv", delimiter=",") 
```

```python
pl.plot(adv1[:,0], adv1[:,1], ls = "--", label="Pe = Infinite") 
pl.scatter(diff1[:,0], diff1[:,1],marker ="o", facecolors='none', label="Pe = 320") 
pl.scatter(diff2[:,0], diff2[:,1],marker ="s", label="Pe = 80" )       
pl.scatter(diff3[:,0], diff3[:,1],marker ="*", label="Pe = 8")     
pl.xlim(0,2.5) 
pl.ylim(0, 1.05)   
pl.legend(loc = "lower right")  
plt.xlabel(r'Pore volume [ ]',fontsize=12)
pl.ylabel(r' Normalized concentration, $\frac{C}{C_0}$ [ ] ')          
```

```python
fig2 = np.genfromtxt('figf2.csv', delimiter=',')
pl.loglog(fig2[:,0], fig2[:,1]) 
pl.xlim(10**-3, 100) 
pl.ylim(0.1, 100) 
pl.axvline(x=fig2[3,0], ymin=0.5, ymax=0.01, ls="--")
pl.axvline(x=fig2[11,0], ymin=0.8, ymax=0.001, ls="--")       
 
pl.annotate("Diffusion", (0.004, 10), (0.002, 0.2), fontsize=12) 
pl.annotate("Control", (0.004, 10), (0.002, 0.13), fontsize=12) 
pl.annotate("Dispersion", (0.004, 10), (0.05, 0.3), fontsize=12) 
pl.annotate("diffusion", (0.004, 10), (0.05, 0.2), fontsize=12) 
pl.annotate("Control", (0.004, 10), (0.05, 0.13), fontsize=12)
pl.annotate("Advection", (0.004, 10), (6, 0.3), fontsize=12) 
pl.annotate("dispersive", (0.004, 10), (6, 0.2), fontsize=12) 
pl.annotate("Control", (0.004, 10), (6, 0.13), fontsize=12) 
pl.xlabel(r'$Pe = \frac{v_x d}{D_p}$', fontsize=18) 
pl.ylabel(r'$\frac{D_L}{D_p}$', fontsize=18) 
pl.grid(which="major", axis="y")    
pl.savefig("fig2.pdf") 
```
